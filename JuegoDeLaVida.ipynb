```
class GameOfLife:

  CelViva = 1
  CelMuert = 0

  def __init__(self, ren, col, gen, pob):
    self.largo = col
    self.alto = ren
    self.grid = Array2D(self.alto, self.largo, 0)
    self.grid.clearing(self.CelMuert)
    self.gener = gen

    for cel in pob:
      self.grid.set_item(cel[0], cel[1], self.CelViva)
  
  def imp_grid(self):
    for r in range(self.grid.get_num_rows()):
            for c in range(self.grid.get_num_cols()):
                if self.grid.get_item(r,c) == 0:
                    print("░░",end="")
                else:
                    print("▓▓",end="")
            print("")

  def conf_gen(self, pob_nueva):
    self.grid.clearing()
    for celu in pob_nueva:
      self.grid.set_item(celu[0], celu[1], self.CelViva)
  
  def get_num_rows(self):
    return self.alto

  def get_num_cols(self):
    return self.largo

  def set_cel_muert(self, rens, colum):
    self.grid.set_item(rens, colum, self.CelMuert)

  def set_cel_viva  (self, rens, colum):
    self.grid.set_item(rens, colum, self.CelViva)

  def get_is_muerta(self, rens, colum):
    return self.grid.get_item(rens, colum) == self.CelMuert
  
  def get_is_viva(self, rens, colum):
    return self.grid.get_item(rens, colum) == self.CelViva

  def get_num_vec(self, rens, colum):
        limt = self.calc_vec(rens, colum)
        contVec = 0
        for f in range(limt[0], limt[2]+1):
            for c in range(limt[1],limt[3]+1):
                if f == rens and c == colum:
                    pass
                else:
                    if self.get_is_viva(f, c):
                        contVec = contVec + 1
        return contVec

  def calc_vec(self, b, a):
        vec = [ b-1, a-1, b+1, a+1]
        if vec[0] == -1:
            vec[0] == 0
        if vec[1] == -1:
            vec[1] == 0
        if vec[2] == self.alto :
            vec[2] = self.alto -1
        if vec[3] == self.largo:
            vec[3] = self.largo -1
        return vec

  def evol(self):
    self.imp_grid()
    c = 0
    while c <= self.gener:
      c = c + 1
      print("\n")
      pob_n = []
      pob_n.clear()
      pob_n = []
      pob_n.clear()
      for row in range(self.grid.get_num_rows()):
        for col in range(self.grid.get_num_cols()):
          if self.grid.get_item(row,col) == self.CelMuert:
            elemento = self.get_num_vec(row,col)
            if elemento == 3:
              pob_n.append((row, col))
            else:
              continue
          elif self.grid.get_item(row, col) == self.CelViva:
            elemento = self.get_num_vec(row,col)
            if elemento == 2:
              pob_n.append((row, col))
            elif elemento == 3:
             pob_n.append((row, col))
            else:
              continue
      self.conf_gen(pob_n)
      self.imp_grid()

inic = GameOfLife(7, 7, 5, [(1,2), (2,1), (2,2), (2,3)])
inic.evol()

  
```
