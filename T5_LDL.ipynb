```
class NodoDoble:
   def __init__( self , value , previo= None , sig= None ):
        self.data = value
        self.prev = previo
        self.next = sig

class DoubleLinkedList:
  def __init__(self):
    self.__head = None
    self.__tail = None
    self.__size = 0 

  def get_size(self):
    return self.__size

  def is_empty(self):
    return self.__size == 0

  def append(self, value):
    if self.is_empty():
      nuevo = NodoDoble(value)
      self.__head = nuevo
      self.__tail = nuevo
    else:
      nuevo = NodoDoble(value, self.__tail, None )
      self.__tail.next = nuevo
      self.__tail = nuevo

    self.__size += 1

  def transversal(self):
    curr_node = self.__head
    while curr_node != None:
      print(" <---", curr_node.data, "---> ", end="")
      curr_node = curr_node.next
    print("")

  def reverse_transversal(self):
    curr_node = self.__tail
    while curr_node != None:
      print(" <---", curr_node.data, "---> ", end="")
      curr_node = curr_node.prev
    print("")  

  def find_from_head(self, value):
    curr_node = self.__head
    cont = 0
    while curr_node != None:
      cont += 1
      if value == curr_node.data:
        break
      curr_node = curr_node.next
    return cont

  def find_from_tail(self, value):
    curr_node = self.__tail
    cont = 0
    while curr_node != None:
      cont += 1
      if value == curr_node.data:
        break
      curr_node = curr_node.prev
    return cont

  def remove_from_head(self, value):
    curr_node = self.__head
    while curr_node != None:
      if value == self.__head.data:
        self.__head = self.__head.next
        self.__head.prev = None
        curr_node.next = curr_node
        break
      if curr_node.next.data == value and curr_node.next != None:
        ant = curr_node
        ant.next = curr_node.next.next
        ant.next.prev = ant
        break
      if value == self.__tail.data and curr_node == self.__tail:
        self.__tail = self.__tail.prev
        self.__tail.next = None       
      curr_node = curr_node.next

  def remove_from_tail(self, value):
    curr_node = self.__tail
    while curr_node != None:
      if value == self.__tail.data:
        self.__tail = self.__tail.prev
        self.__tail.next = None
        curr_node.prev = curr_node
        break
      if curr_node.prev.data == value and curr_node.prev != None:
        ant = curr_node
        ant.prev = curr_node.prev.prev
        ant.prev.next = ant
        break
      if value == self.__head.data and curr_node == self.__head:
        self.__head = self.__head.next
        self.__head.prev = None       
      curr_node = curr_node.next

```
